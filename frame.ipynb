```python
class Frame:
    def __init__(self, category_target, current_features=None):
        if current_features is None:
            current_features = []

        self.category_target = category_target
        self.current_features = set(current_features)
        self.is_completed = False
        self.mood = random.randint(0,3)
        self.personaggio = "Obi Wan Kenobi"
        
        """
        if self.personaggio == 'Darth Fener':
            self.mood = random.randint(0,2) #il mood di Darth Fener è random
        
        elif self.personaggio == 'Obi Wan Kenobi':
            self.mood == 0 #il mood di Obi Wan Kenobi è sempre 0
        """
    
    def get_mood(self):
        return self.mood
    
    def set_mood(self, mood):
        self.mood = mood

    def get_personaggio(self):
        return self.personaggio
    
    def set_personaggio(self, personaggio):
        self.personaggio = personaggio
    
    def setting_mood(self, mood):
        if self.personaggio == 'Darth Fener':

            mood = random.randint(0,3) #il mood di Darth Fener è random

        if self.personaggio == 'Obi Wan Kenobi':
            mood = 0
        
        return mood
    

    def get_category_target(self):
        return self.category_target
    
    def get_list_features(self):
        return self.current_features

    #controllo se la mia feature è presente nella lista della categoria
    def get_target_list_features(self):
        features = self.get_category_target().get_features()
        result = []
        for f in features:
            result.append(f.lower())

        return result

    #controllo se posso aggiungere un elemento. In questo caso controllo se la feature aggiunta e' quella corretta o no
    def add_element(self, feature):
        feature = set(feature)
        for f in feature:
            self.check_correct_feature(f)

    def check_correct_feature(self, feature):
        found = False
        #controllo se la mia feature e' corretta
        for target_feature in self.category_target.get_features():
            if feature.lower() == target_feature.lower():
                self.current_features.add(feature.lower())
                found = True
        
        #print("SET:", set(self.get_target_list_features()))
        if not found:
            print(f"Feature non corretta: {feature}")


    #controllo se il numero di feature presenti nella lista di una categoria sono state inserite tutte
    def check_finish(self):
        count = 0
        for f in set(self.get_target_list_features()):
            if f in self.current_features:
                count += 1
        
        if count == len(set(self.get_target_list_features())): #se il numero di feature corrette e' uguale al numero di feature totali
            self.is_completed = True
            #print(f"Hai terminato la lista di caratteristiche per la categoria {self.get_category_target().get_name()}")
        
        else:
            #print(f"Non hai ancora terminato la lista di caratteristiche per la categoria {self.get_category_target().get_name()}")
            self.is_completed = False
        
        return self.is_completed

if __name__ == "__main__":
    #test
    cat = Frame(jedi_order)
    #print(cat)
    print(cat.get_category_target())
    cat.add_element(["Force"])
    print(cat)

    #ok non aggiunge nulla 
    cat.add_element(["Agricultural"])
    
    print("Finish: ", cat.check_finish())
    #ok aggiunge
    cat.add_element([ 
                    "self-discipline",
                    "knowledge"]
                    )
    #ok check corretto
    print(f"Finish: ", cat.check_finish())
    #print(f"cat: {cat.get_category_target()}")
```
