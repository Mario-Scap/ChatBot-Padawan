```python
class Categories:
    def __init__(self, name, questions, features, negative_features):
        self.name = name
        self.questions = questions
        self.features = features
        self.negative_features = negative_features


    #Name of the rank 
    def get_name(self):
        return self.name
    
    #Name of the rank 
    def get_questions(self):
        return self.questions
    
    def get_score(self):
        return self.score
    
    def get_features(self):
        return self.features
    
    def get_negative_features(self):
        return self.negative_features

    def __str__(self):
        return f'{self.name}\n\tName List: {", ".join(self.features)}'
    
```

```python
class Question:
    def __init__(self, question, answer):
        self.question = question
        self.answer = answer
        self.user_answer = ""
        self.already_asked = False
    
    def get_question(self):
        return self.question
    
    def get_answer(self):
        return self.answer
    
    def set_user_answer(self, user_answer):
        self.user_answer = user_answer

    def set_asked(self):
        self.already_asked = True
    
    def check_asked(self):
        return self.already_asked
```
